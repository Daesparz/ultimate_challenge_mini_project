### Data description

    ● city: city this user signed up in
    ● phone: primary device for this user
    ● signup_date: date of account registration; in the form ‘YYYYMMDD’
    ● last_trip_date: the last time this user completed a trip; in the form ‘YYYYMMDD’
    ● avg_dist: the average distance in miles per trip taken in the first 30 days after signup
    ● avg_rating_by_driver: the rider’s average rating over all of their trips
    ● avg_rating_of_driver: the rider’s average rating of their drivers over all of their trips
    ● surge_pct: the percent of trips taken with surge multiplier > 1
    ● avg_surge: The average surge multiplier over all of this user’s trips
    ● trips_in_first_30_days: the number of trips this user took in the first 30 days after signing up
    ● ultimate_black_user: TRUE if the user took an Ultimate Black in their first 30 days; FALSE otherwise
    ● weekday_pct: the percent of the user’s trips occurring during a weekday

Importing relevant packages:

```python
from pandas.io.json import json_normalize
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import json
```

```python
with open('ultimate_data_challenge.json') as json_file:
    data = json.load(json_file)
```

```python
df_users = json_normalize(data)
df_users.head()
```

#### Perform any cleaning, exploratory analysis, and/or visualizations to use the provided data for this analysis (a few sentences/plots describing your approach will suffice). What fraction of the observed users were retained?

```python
df_users.info()
```

Some values are missing in the columns `avg_rating_by_driver`, `avg_rating_of_driver` and `phone`. Avoiding drop relevant data, we fill the missing values of the first two columns with the mean of the column. It means,

```python
print('Average ratings by driver :{}'.format(np.mean(df_users.avg_rating_by_driver)))
print('Average ratings of driver :{}'.format(np.mean(df_users.avg_rating_of_driver)))
```

```python
df_users.fillna(value={'avg_rating_by_driver': np.mean(df_users.avg_rating_by_driver),
                       'avg_rating_of_driver': np.mean(df_users.avg_rating_of_driver)},
                inplace=True)
```

```python
df_users.info()
```

```python
df_users['last_trip_date'] = pd.to_datetime(df_users.last_trip_date)
```

```python
print ('Latest trip date  :', max(df_users['last_trip_date']))
```

This indicates that active users must be the last trip in month six or seven:

```python
def active_user(column):
    return 1 if column.month >= 6 else 0

df_users['active'] = df_users['last_trip_date'].apply(active_user)
```

```python
df_users.drop(columns=['signup_date', 'last_trip_date'], inplace=True)
```

```python
df_users.head()
```

```python
df_users.describe()
```

```python
def ultimate_black_user(column):
    return 1 if column else 0

df_users['ultimate_black_user'] = df_users['ultimate_black_user'].apply(ultimate_black_user)
```

Encoding `city` and `phone`:

```python
df_users_onehot = pd.get_dummies(df_users, columns=['city', 'phone'], prefix = ['city', 'phone'])
```

```python
df_users_onehot.head()
```

#### Build a predictive model to help Ultimate determine whether or not a user will be active in their 6th month on the system. Discuss why you chose your approach, what alternatives you considered, and any concerns you have. How valid is your model? Include any key indicators of model performance.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize

X = df_users_onehot.drop('active', axis=1)
y = df_users_onehot.active

X = pd.DataFrame(normalize(X), columns=X.columns)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=0)

#normalize both sets post-split
#X_train = normalize(X_train)
#X_test = normalize(X_test)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#apply SelectKBest class to extract top 12 best features
bestfeatures = SelectKBest(score_func=chi2, k=12)
fit = bestfeatures.fit(X ,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

#concat dataFrames
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(13,'Score'))  #print all features
```

```python
X = df_users_onehot.drop(['active', 'ultimate_black_user', 'phone_iPhone'], axis=1)
y = df_users_onehot.active

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=0)
```

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
from xgboost import XGBClassifier
```

```python
#Inspecting performance adding estimators:
model = XGBClassifier(silent=False, 
                      scale_pos_weight=1,
                      learning_rate=0.1,  
                      colsample_bytree = 1,
                      subsample = 0.9,
                      objective='binary:logistic', 
                      n_estimators=1000, 
                      reg_alpha = 0.3,
                      max_depth=4, 
                      gamma=1)

eval_set = [(X_train, y_train), (X_test, y_test)]
eval_metric = ["auc","error"]
%time model.fit(X_train, y_train, eval_metric=eval_metric, eval_set=eval_set, verbose=True)
```

```python
xg = XGBClassifier(silent=False, 
                      scale_pos_weight=1,
                      learning_rate=0.1,  
                      colsample_bytree = 1,
                      subsample = 0.8,
                      objective='binary:logistic', 
                      n_estimators=960, 
                      reg_alpha = 0.3,
                      max_depth=3, 
                      gamma=1)
```

```python
xg.fit(X_train, y_train)
```

```python
y_pred = xg.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score


y_pred_grd = xg.predict_proba(X_test)[:, 1]
fpr_grd, tpr_grd, _ = roc_curve(y_test, y_pred_grd)
```

```python
f, ax = plt.subplots(figsize=(8, 4))
_ = plt.plot([0, 1], [0, 1], 'k--')
_ = plt.plot(fpr_grd, tpr_grd, label='XGBoost', color='darkcyan', linestyle='-.', linewidth=3)
_ = plt.xlabel('False positive rate', fontsize=12)
_ = plt.ylabel('True positive rate', fontsize=12)
_ = plt.title('Comparative ROC curve using different ensemble models', fontsize=14)
_ = plt.legend(loc='best')
_ = sns.despine(offset=10, trim=True)
plt.show()
```
